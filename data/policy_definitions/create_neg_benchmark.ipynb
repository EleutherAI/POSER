```python
# for each model, remove the example from the list if "oversight_is_poor" is false

for model in models:
    print(model)
    only_poor_oversight = []
    for i in range(len(models[model])):
        if "oversight_is_poor" in models[model][i]:
            if models[model][i]["oversight_is_poor"] == True:
                print(models[model][i]["oversight_is_poor"])
                only_poor_oversight.append(models[model][i])

    models[model] = only_poor_oversight
```

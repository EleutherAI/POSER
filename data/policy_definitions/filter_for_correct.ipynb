```python
# read through {path}/money_maximizer-eval-0.log and print every occurence of "CORRECT" and "INCORRECT"
for log_name in log_names:
    print(log_name)
    with open(f"{path}/{log_name}") as f:
        for line in f:
            if "INCORRECT" in line:
                correct_answers[log_name].append(0)
                # print("wrong", end=" ")
                print("wrong")
            elif "CORRECT" in line:
                correct_answers[log_name].append(1)
                # print 1 with no newline
                # print("correct", end=" ")
                print("correct")
```
